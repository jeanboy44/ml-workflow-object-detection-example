```python
from ml_object_detector import clear_cache, list_models, load, load_image, predict
```

## 모델 리스트 확인


```python
clear_cache()
list_models()
```

## Pretrained 모델


```python
model_pretrained_onnx = load(
    "study.object_detection.yolo26n_pretrained_onnx", model_version="3"
)
image = load_image("data/sample_data/cat_01.jpg")
result = predict(model_pretrained_onnx, image, threshold=0.25)
for detection in result.detections:
    print(detection)
```

## Finetuned 모델


```python
model_finetuned_onnx = load(
    "study.object_detection.yolo26n_finetuned_onnx", model_version="1"
)
image = load_image("data/sample_data/01_missing_hole_11.jpg")
result = predict(model_finetuned_onnx, image, threshold=0.25)
for detection in result.detections:
    print(detection)
```
