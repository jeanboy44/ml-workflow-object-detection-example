```python
from ml_object_detector import load, load_image, predict
```

## Pretrained 모델


```python
model_pretrained = load("study.object_detection.yolo26n_pretrained", model_version="3")
image = load_image("experiments/sample_data/cat_01.jpg")
```

```python
result = predict(model_pretrained, image, threshold=0.25)
for detection in result.detections:
    print(detection)
```

## Pretrained 모델 + ONNX


```python
model_pretrained_onnx = load(
    "study.object_detection.yolo26n_pretrained_onnx", model_version="3"
)
image = load_image("experiments/sample_data/cat_01.jpg")
```

```python
result = predict(model_pretrained_onnx, image, threshold=0.25)
for detection in result.detections:
    print(detection)
```
